3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def normalize_matrix(matrix):
    return matrix / np.sum(matrix)

A = np.array([[1, 2], [3, 4]])

print(normalize_matrix(A))
```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def pyramid_volume(x, y, z):
    return abs(np.dot(x, np.cross(y, z))) / 6

v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

print(pyramid_volume(v1, v2, v3))
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def border_matrix(N):
    mat = np.ones((N, N), dtype=int)
    mat[1:-1, 1:-1] = 0
    return mat

N = 3
print(border_matrix(N))
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
plt.figure(figsize=(8, 8))
cases = merged_data.groupby('continent')['total_cases'].sum()
cases.plot(
    kind='pie',
    autopct=lambda p: f"{int(p * cases.sum()/100):,}",
    startangle=90
)
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
countries = ['Russia', 'China', 'United States', 'India', 'Nigeria']
tests_data = merged_data[merged_data['location'].isin(countries)].sort_values('total_tests', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(
    x='location',
    y='total_tests',
    data=tests_data,
    hue='location',
    palette='viridis',
    legend=False
)
plt.title('Число тестов на COVID-19 по странам')
plt.xlabel('Страны')
plt.ylabel('Общее количество тестов')
plt.tight_layout()
plt.show()
```

3. Визуализируйте зависимость ВВП страны и случаев заболевания COVID-19

```python
plot_data = merged_data.dropna(subset=['gdp_per_capita', 'total_cases'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp_per_capita', y='total_cases', data=plot_data, hue='continent')
plt.xscale('log')
plt.yscale('log')
plt.title('Зависимость между ВВП на душу и случаями COVID-19')
plt.xlabel('ВВП на душу')
plt.ylabel('Общее число случаев')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

4. Постройте гистограмму распределения количества смертей на один миллион человек

```python
plt.figure(figsize=(10, 6))
sns.histplot(merged_data['deaths_per_million'], bins=30, kde=True)
plt.title('Распределение количества смертей на 1 млн человек')
plt.xlabel('Смертей на 1 млн населения')
plt.ylabel('Количество стран')
plt.show()
```

5. Постройте диаграмму отображающую среднюю продолжительность жизни в разных странах европы

```python
europe_data = merged_data[merged_data['continent'] == 'Europe'].sort_values('life_expectancy')

plt.figure(figsize=(12, 8))
sns.barplot(x='life_expectancy', y='location', data=europe_data, palette='coolwarm', hue='location', legend=False)
plt.title('Средняя продолжительность жизни в странах Европы')
plt.xlabel('Продолжительность жизни')
plt.ylabel('Страна')
plt.tight_layout()
plt.show()
```

6. Сравните распределения кол-ва смертей на 1000 человек в Европе и Африке

```python
plt.figure(figsize=(12, 8))
eur_deaths = merged_data[merged_data['continent'] == 'Europe']['deaths_per_million'] / 1000
afr_deaths = merged_data[merged_data['continent'] == 'Africa']['deaths_per_million'] / 1000

sns.histplot(eur_deaths, bins=20, color='green', kde=True, alpha=0.4, label='Европа')
sns.histplot(afr_deaths, bins=20, color='red', kde=True, alpha=0.4, label='Африка')

plt.title('Смертность от COVID-19 на 1000 человек')
plt.xlabel('Смертей на 1000 человек')
plt.ylabel('Количество стран')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
